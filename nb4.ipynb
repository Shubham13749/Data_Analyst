```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as tsp
import seaborn as snc
```

```python
df = pd.read_csv('./death per year by continent.csv')
print(df.tail(10))
```

```python
df.drop(['Latin America and the Caribbean (UN)','Northern America (UN)','Unnamed: 8'], axis=1,inplace=True)
```

```python
tdf=df.tail(10)
tdf
```

```python
snc.barplot(data=tdf, x=tdf.Africa, y=tdf.Years, hue="World",palette='Accent_r')
tsp.show()
tsp.grid()
tsp.title("Death-Rate")
```

```python
snc.scatterplot(data=tdf, x=tdf.Years, y=tdf.Africa, hue=tdf.World,style="World",size=1000,
                palette='Accent_r',markers=['o','>'])
tsp.grid()
tsp.title("Death-Rate")
tsp.show()
```

```python
import pandas as pd

Data={
    'Subject':['Maths','Science','English','Social science','Hindi'],
    'Marks':[85,75,99,85,99]
}
df=pd.DataFrame(Data)
df
```

```python
snc.lineplot(data=df,x=df.Subject,y=df.Marks,hue='Marks',style='Marks',palette='Accent')
tsp.title("PCS CLASS 10TH RESULT")
tsp.show()
```

```python
order_1 =['Maths','Science','English','Social science','Hindi']
snc.barplot(data=df,x=df.Subject,y=df.Marks,hue='Marks',palette='crest',saturation=10,order=order_1)
tsp.grid()
tsp.title("PCS CLASS 10TH RESULT")
tsp.show()
```

```python
snc.scatterplot(data=df,x=df.Subject,y=df.Marks,hue='Marks',style='Marks',palette='flare')
tsp.grid()
tsp.title("PCS CLASS 10TH RESULT")
tsp.show()
```

```python
snc.distplot(df['Marks'],kde=True,rug=True,bins=[50,60,70,80,90,100,110],color='g')
tsp.grid()
tsp.title("PCS CLASS 10TH RESULT")
tsp.show()
```
